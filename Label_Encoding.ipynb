```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
from google.colab import files
import io
upoladed=files.upload()
```

```python
df=pd.read_csv(io.BytesIO(upoladed['breast_cancer_data.csv']))
```

```python
df.head()
```

```python
df
```

```python
df["diagnosis"].value_counts() #count the differnt types in columns
```

```python
label_encode=LabelEncoder() # loading the label encoder and giving him name label_encode
```

```python
labels=label_encode.fit_transform(df['diagnosis'])
```

```python
df['target']=labels
```

```python
df['target'].value_counts()
```
